```python
import pandas as pd


# Create a list to store the data split into rows
data_lines = []

# Read data from the file
with open('../data/raw-data/newsCorpora.csv', 'r', encoding='utf-8') as file:
    # Split the data into lines and store them in the list
    for line in file:
        data_lines.append(line.strip().split("\t"))

# Create dataframe from lines
df = pd.DataFrame(data_lines)

# Assign columns header
df.columns = ["ID", "TITLE", "URL", "PUBLISHER", "CATEGORY", "STORY", "HOSTNAME", "TIMESTAMP"]

df
```

```python
# Extract article from publishers of interest
selected_publishers = ["Reuters", "Huffington Post", "Businessweek", "Contactmusic.com", "Daily Mail"]
df = df[df["PUBLISHER"].isin(selected_publishers)].copy().reset_index(drop=True)
 
df
```

```python
# Select only the title and category for classification
df = df[["CATEGORY", "TITLE"]].copy()
df
```

```python
def clean_sentence(sentence):    
    """
    Clean a sentence by removing special symbols, whitespace errors,
    and lower all characters.

    Parameters:
    sentence (str): The input sentence to be cleaned.

    Returns:
    str: The cleaned sentence.

    Note:
    The function is designed to be applied to article titles, which are assumed
    to be relatively clean already. Thus, it focuses on removing unnecessary 
    elements and normalizing the text for further processing. 
    """
    
    # Remove special symbols
    for symbol in ['*', '?', '%', '&', '$', '(', ')', '#', '^', '@', '!', '~', '-', 'â€”', '+', '=', ",", "'", '"',"/", ".", "-", ":", ";", ">", "<"]:
        if symbol in sentence:
            sentence = sentence.replace(symbol, ' ')
        
    # Convert all text to lowercase
    sentence = sentence.lower()

    # Remove whitespace errors
    sentence = ' '.join([word.strip() for word in sentence.split()])
    
    return sentence

# Apply sentence cleaning for title
df['TITLE'] = df['TITLE'].apply(clean_sentence)

# Shuffle the example 
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

df
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print("train set size: ", len(train_df))
print("valid set size: ", len(valid_df))
print("test set size: ", len(test_df))

# save all the sets to file
train_df[['CATEGORY', 'TITLE']].to_csv('../data/cleaned-data/train.txt', encoding='utf-8', sep='\t', index=False, header=False)
valid_df[['CATEGORY', 'TITLE']].to_csv('../data/cleaned-data/valid.txt', encoding='utf-8', sep='\t', index=False, header=False)
test_df[['CATEGORY', 'TITLE']].to_csv('../data/cleaned-data/test.txt', encoding='utf-8', sep='\t', index=False, header=False)
```
